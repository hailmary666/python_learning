```python
import matplotlib.pyplot as plt
```

```python
country = df[df['Country'] == 'Russia']
```

```python
print(f"founded: {country.shape[0]}")
print(country)
```

```python
print(df.columns.tolist())
```

```python
years = ['2022 Population', '2020 Population', '2015 Population', '2010 Population', '2000 Population', '1990 Population', '1980 Population', '1970 Population']
years.reverse()
```

```python
population_values = country[years].values.flatten()
```

```python
plt.plot(years, population_values, marker='s')
plt.title('Population of Russia from 1970 to 2022')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

```python
df_sorted = df.sort_values('2022 Population', ascending=False).head(5)
```

```python
plt.bar(df_sorted['Country'], df_sorted['2022 Population'])
```

```python
plt.scatter(df['2000 Population'], df['2022 Population'], s=0.1)
plt.grid(True)
```

Этот scatter plot показывает взаимосвязь между населением стран в 2000 и 2022 годах. Большинство точек расположено близко к воображаемой линии, идущей из нижнего левого угла в верхний правый. Это говорит о сильной прямой зависимости: страны с большим населением в 2000 году, как правило, сохранили его к 2022 году. Однако есть несколько точек-выбросов: 1) Страны, у которых значение по оси Y (2022) значительно ниже — это, вероятно, государства, пережившие резкую убыль населения (например, из-за войн или экономического кризиса). 2) Страны, у которых значение по оси Y значительно выше — возможно, это страны с очень высоким приростом населения или те, чьи границы/методы учёта изменились. Для проверки этих гипотез нужно выделить эти выбросы и посмотреть их названия.
